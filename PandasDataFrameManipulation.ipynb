Pandas DataFrame Manipulation

```python
import pandas as pd
data = {'Name': ['Abu Bakar', 'Bilal', 'Hassan', 'Alex', 'Mitchel Starc'],
        'Age': [25, 22, 17, 32, 35],
        'City': ['Berlin', 'Karachi', 'Hamburg', 'Berlin', 'Sydney']}
df = pd.DataFrame(data)
print(df)
```

Add a New Column to a Pandas DataFrame

```python
country = ["Germany", "Pakistan", "Germany", "Germany", "Austrailia"]
df["Country"] = country
```

```python
df
```

Add a New Row to a Pandas DataFrame

```python
# As we dont know the last index and we want to add at the end of the data a new row
df.loc[len(df.index)] = ["Habib Ullah" , 26, "London", "England"]
```

```python
df
```
