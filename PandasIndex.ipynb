Pandas Index

Default Index
When we create a DataFrame or Series without specifying an index explicitly, Pandas assigns a default integer index starting from 0.

```python
import pandas as pd
data  = [
    ["Abu Bakar",25,"Berlin, Germany"],
    ["Bilal",22,"Karachi, Pakistan"],
    ["Hassan", 17,"Hamburg, Germany"]
]
Df = pd.DataFrame(data, columns = ["Name", "Age", "City"])
print(Df)
```

Setting Index

```python
data2 = {
    'Name' : ["Uzair","Bilal","Hassan"],
    'Age' : [25,22,17],
    'City' : ["Berlin, Germany", "Karachi, Pakistan", "Hamburg, Germany"]
}
Df2 = pd.DataFrame(data2)
print(Df2)

# set the 'Name' column as index
Df2.set_index('Name', inplace=True)
print("After Modification: \n ", Df2)
```

```python
data3 = {
    'Name' : ["Uzair","Bilal","Hassan"],
    'Age' : [25,22,17],
    'City' : ["Berlin, Germany", "Karachi, Pakistan", "Hamburg, Germany"]
}
Df3 = pd.DataFrame(data3, index= pd.RangeIndex(5,8, name = "Index"))

print("After Modification: \n ", Df3)
```

```python
import pandas as pd
data  = [
    ["Abu Bakar",25,"Berlin, Germany"],
    ["Bilal",22,"Karachi, Pakistan"],
    ["Hassan", 17,"Hamburg, Germany"]
]
Df = pd.DataFrame(data, columns = ["Name", "Age", "City"])
print("Original DataFrame With Default Index: \n", Df)

Df.rename(index = {0:"A", 1:"B", 2:"C"}, inplace= True)

print("After Modification: \n ", Df)
```

reseting the indexes

```python
Df.reset_index(inplace= True)
print("After Resetting: \n", Df)
```

Access Rows by Index

```python
import pandas as pd

# create a dataframe
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 28, 32],
        'City': ['New York', 'London', 'Paris']}

df = pd.DataFrame(data)
print(df)
secondRow = df.iloc[1]

print("Second Row: \n",secondRow)
```
