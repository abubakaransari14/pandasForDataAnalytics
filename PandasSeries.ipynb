Pandas is a Python library used for data manipulation and analysis. Pandas provides a convenient way to analyze and clean data.

Pandas Series

```python
import pandas as pd

data = [10, 20, 30, 40, 50]
mySeries = pd.Series(data)

print(mySeries)
```

```python
mySeries[2]
```

```python
Data = ["Abu Bakar","Uzair","Umair","Ubaid","Bilal","Hassan"]
print(Data)
# create a series and specify labels
studentData = pd.Series(Data, index = ["001","011","033","022","002","003"])
print(studentData)
```

```python
studentData["001"]
```

```python
#Create Series From a Python Dictionary
grades = {"Abu Bakar": 3.26, "Bilal": 3.11, "Hassan": 3.33}
myStudentGrades = pd.Series(grades)
print(myStudentGrades)
```

```python
myStudentGrades["Abu Bakar"]
```
