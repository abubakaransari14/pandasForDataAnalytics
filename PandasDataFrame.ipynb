Pandas DataFrame

Pandas DataFrame Using Python Dictionary

```python
import pandas as pd


data = {'Name': ['Abu Bakar', 'Bilal', 'Hassan'],
       'Age': [25, 22, 17],
       'City': ['Berlin', 'Karachi', 'Hamburg']}

# create a dataframe from the dictionary
df = pd.DataFrame(data)

print(df)
```

Pandas DataFrame Using Python List

```python
import pandas as pd

# create a two-dimensional list
data = [['Abu Bakar', 25, 'Berlin'],
       ['Bilal', 22, 'Karachi'],
       ['Hassan', 17, 'Hamburg']]

df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])

print(df)
```
