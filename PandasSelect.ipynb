Pandas Select

Select Data Using Indexing and Slicing

```python
import pandas as pd
data = {'Name': ['Abu Bakar', 'Bilal', 'Hassan', 'Alex', 'Mitchel Starc'],
        'Age': [25, 22, 17, 32, 35],
        'City': ['Berlin', 'Karachi', 'Hamburg', 'Berlin', 'Sydney']}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Name'] #selecting single column name
```

```python
df[['Name','Age']]
```

```python
# selecting multiple columns
age_salary_columns = df[['Age', 'City']]

print("Selecting multiple columns: Age and City")
print(age_salary_columns.to_string(index=False))
print()
```

```python
# selecting rows using slicing
df[0:2]
```

Using loc and iloc to Select Data

```python
df.loc[:,['Name','City']]
```

```python
df.iloc[0:3,0:2]
```

```python
df.iloc[[0,2],[1,0]]
```

```python
df.loc[[0, 2], ['Name', 'City']]  #(.loc[] does NOT support slicing with lists.)
```

Select Rows Based on Specific Criteria

```python
df[df['Name'] == 'Abu Bakar']
```

```python
# df[df['Name'].str.startswith('A', case=False)]
df[df['Name'].str.startswith('A')]
```

```python
# df.loc[(df['Name'].str.startswith('A')) & (df['Age']>26)]
df[(df['Name'].str.startswith('A')) & (df['Age']>26)]
```

```python
df[(df['Name'].str.contains('A', case= False)) & (df['Age']>24)]
```

query() to Select Data

```python
df.query('Age >= 24')
```

```python
df.query('Age >= 24')
```

```python
nameFiltered = ['Abu Bakar','Bilal']
df[df['Name'].isin(nameFiltered)]
```
