Pandas DataFrame Analysis

Pandas head()

```python
import pandas as pd

# create a dataframe
data = {'Name': ['Abu Bakar', 'Bilal', 'Hassan', 'Zain', 'Alex', 'Memo', 'Siri', 'Aman', 'Tom', 'Eva'],
        'Age': [25, 22, 17, 28, 32, 27, 40, 33, 29, 31],
        'City': ['Berlin', 'Karachi', 'Hamburg', 'Sydney', 'Berlin', 'Berlin', 'Rome', 'Madrid', 'Toronto', 'Moscow']}
df = pd.DataFrame(data)
print(df)
print()
print("Using head(3) function: ")
print()
print("In head(3) it will show first 3 row: \n",df.head(3))
print()
print("If with head() it will default show 1st 5 row: \n",df.head())
```

The tail() method is similar to head() but it returns data starting from the end of the DataFrame.

```python
print('Last Three Rows:')
print(df.tail(3))
print()

# display the last five rows
print('Last Five Rows:')
print(df.tail())
```

The info() method gives us the overall information about the DataFrame such as its class, data type, size etc.

```python
# get info about dataframe
df.info()
```
