Access Columns of a DataFrame

```python
import pandas as pd
data = {'Name': ['Abu Bakar', 'Bilal', 'Hassan', 'Alex', 'Mitchel Starc'],
        'Age': [25, 22, 17, 32, 35],
        'City': ['Berlin', 'Karachi', 'Hamburg', 'Berlin', 'Sydney']}
df = pd.DataFrame(data)
```

.loc vs .iloc

Pandas .loc

In Pandas, we use the .loc property to access and modify data within a DataFrame using label-based indexing.

It allows us to select specific rows and columns based on their labels.

Syntax
The syntax of .loc in Pandas is:

```python
# access a single row
singleRow = df.loc[2]

print("Single row:")
print(singleRow)
print()
```

```python
# access rows 0, 3 and 4
rowList = df.loc[[0, 3, 4]]

print("List of Rows:")
print(rowList)
print()
```

```python
df.loc[:,['Name','City']]
```

```python
 df.loc[1:3]
```

```python
 df.loc[:, "Name":"Age"]
```

```python
df.loc[df["Age"]>30]
```

```python
df.iloc[1]
```

```python
df.iloc[1:3]
```

```python
df.iloc[:,[0,2]]
```

```python
df.iloc[:,[0,2]]
```

```python
df.iloc[[1,4],:]
```
